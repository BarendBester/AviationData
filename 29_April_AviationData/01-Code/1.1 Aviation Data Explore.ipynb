# 1.2 Data Visualisations

### 1.2.1 Look at the strikes over the years

```python
# Group the DataFrame by 'Incident Year' and count the occurrences
strikes_by_year = df_Aviation.groupby('Incident Year').size().reset_index(name='total_strikes')

# Plot the counts for each year
plt.plot(strikes_by_year['Incident Year'], strikes_by_year['total_strikes'], color='blue', linewidth=1)
plt.xlabel('Year')
plt.ylabel('Number of Strikes')
plt.title('Number of Strikes by Year')
plt.xticks([1990, 1995, 2000, 2005, 2010, 2015])
plt.yticks([2000, 6000, 10000, 14000])
plt.grid(False)
plt.show()
```

```python
df_Aviation.info()
```

```python
# Isolate the strikes
strike_columns = ['Radome Strike', 'Windshield Strike', 'Nose Strike', 'Engine1 Strike', 'Engine2 Strike', 
                  'Engine3 Strike', 'Engine4 Strike', 'Propeller Strike', 'Wing or Rotor Strike', 
                  'Fuselage Strike', 'Landing Gear Strike', 'Tail Strike', 'Lights Strike']
strike_data = df_Aviation[strike_columns]
```

```python
# Replace empty strings with NaN
strike_data = strike_data.replace('', pd.NA)
```

```python
# Drop rows with NaN values
strike_data.dropna(inplace=True)
```

```python
# Summarize the strike counts for each part
strike_by_part = strike_data.sum()
```

```python
# Convert the Series to DataFrame and reset index
strike_by_part = strike_by_part.reset_index()
```

```python
# Rename columns
strike_by_part.columns = ['aircraft_part', 'total_number']
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='total_number', y='aircraft_part', data=strike_by_part.sort_values(by='total_number', ascending=False), palette='viridis')
plt.xlabel('Total Number of Strikes')
plt.ylabel('Aircraft Part')
plt.title('Total Number of Strikes by Aircraft Part')
plt.show()
```

```python
# Group the DataFrame by 'Incident Year' and count the occurrences
airstrikes_by_year = df_Aviation.groupby('Incident Year').size().reset_index(name='total_airstrikes')

# Plot the scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(airstrikes_by_year['Incident Year'], airstrikes_by_year['total_airstrikes'], color='blue', s=50)
plt.xlabel('Year')
plt.ylabel('Number of Airstrikes')
plt.title('Airstrikes Over Time')
plt.grid(True)
plt.show()
```

```python
# Select columns of interest
damage_columns = ['Radome Damage', 'Windshield Damage', 'Nose Damage', 'Engine1 Damage', 'Engine2 Damage', 
                  'Engine3 Damage', 'Engine4 Damage', 'Propeller Damage', 'Wing or Rotor Damage', 
                  'Fuselage Damage', 'Landing Gear Damage', 'Tail Damage', 'Lights Damage']
damage_data = df_Aviation[damage_columns]
```

```python
# Replace empty strings with NaN
damage_data = damage_data.replace('', pd.NA)

# Drop rows with NaN values
damage_data.dropna(inplace=True)

# Summarize the damage counts for each part
damage_by_part = damage_data.sum()

# Convert the Series to DataFrame and reset index
damage_by_part = damage_by_part.reset_index()

# Rename columns
damage_by_part.columns = ['aircraft_part', 'total_number']

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='total_number', y='aircraft_part', data=damage_by_part.sort_values(by='total_number', ascending=False), palette='viridis')
plt.xlabel('Total Number of Damage')
plt.ylabel('Aircraft Part')
plt.title('Total Number of Damage by Aircraft Part')
plt.show()
```

### 1.2.4 Number of Species HIT

```python
# Group by species name and count for each species
strikes_by_species = df_Aviation.groupby('Species Name').size().reset_index(name='total_strikes')
```

```python
# This code arranges the strikes_by_species 
# Select top 10
top_species = strikes_by_species.nlargest(10, 'total_strikes')
```

```python
# Plot 
plt.figure(figsize=(10, 6))
sns.barplot(x='total_strikes', y='Species Name', data=top_species, palette='viridis')
plt.xlabel('Total Strikes')
plt.ylabel('Species')
plt.title('Top 10 Species by Total Strikes')
plt.show()
```
